```python
import numpy as np
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, color
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def plot_color_histogram(image):
    # Convertir l'image en espace de couleurs Lab
    lab_image = color.rgb2lab(image)
    
    # Redimensionner l'image pour accélérer le processus de segmentation
    resized_image = lab_image.reshape(-1, 3)
    
    # Utiliser l'algorithme K-means pour segmenter les couleurs
    kmeans = KMeans(n_clusters=4)
    kmeans.fit(resized_image)
    
    # Récupérer les étiquettes des clusters et les compter
    labels = kmeans.labels_
    counts = np.bincount(labels)
    
    # Trouver les couleurs des clusters
    cluster_colors = [color.lab2rgb([[centroid]])[0][0] for centroid in kmeans.cluster_centers_]
    print(cluster_colors)
    # Créer une figure pour afficher l'histogramme des couleurs
    fig, ax = plt.subplots()
    
    # Afficher l'histogramme des couleurs
    ax.bar(range(len(counts)), counts, color=cluster_colors)
    
    # Définir les étiquettes des axes
    ax.set_xlabel('Cluster')
    ax.set_ylabel('Nombre de pixels')
    
    # Afficher la légende avec les couleurs correspondantes
    legend_labels = ['Couleur dominante 1', 'Couleur dominante 2']
    ax.legend(legend_labels, loc='upper right')
    
    # Afficher la figure
    plt.show()

# Charger l'image de la coccinelle
image = io.imread('No_Background_Image/im-004_im.png')

# Afficher l'histogramme des couleurs des élytres
plot_color_histogram(image) 
```

```python
from skimage import io, color
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Charger l'image de la coccinelle
image_path = "im-122_im.png"
image = io.imread(image_path)

# Convertir l'image en espace colorimétrique Lab
image_lab = color.rgb2lab(image)

# Flattening de l'image pour l'entrée de l'algorithme de clustering
image_flat = image_lab.reshape((-1, 3))

# Appliquer le clustering K-means sur l'image
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(image_flat)

# Obtenir les couleurs dominantes
colors_lab = kmeans.cluster_centers_
colors_rgb = color.lab2rgb(colors_lab) * 255

# Trouver le cluster dominant
dominant_cluster = np.argmax(colors_lab[:, 0])

# Créer le masque pour le cluster dominant
mask_dominant_cluster = kmeans.labels_ == dominant_cluster

# Appliquer le masque pour supprimer le fond de l'image
image_no_bg = np.zeros_like(image)
image_no_bg[mask_dominant_cluster.reshape(image.shape[:2])] = image[mask_dominant_cluster.reshape(image.shape[:2])]

# Afficher les couleurs dominantes
plt.figure(figsize=(8, 4))

# Diagramme des couleurs dominantes
plt.subplot(1, 2, 1)
plt.imshow(colors_rgb[np.newaxis, :, :].astype(np.uint8))
plt.title("Couleurs Dominantes")
plt.axis("off")

# Image avec le fond supprimé
plt.subplot(1, 2, 2)
plt.imshow(image_no_bg)
plt.title("Image sans Fond")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Tri des couleurs dominantes par luminosité (canal L)
sorted_indices = np.argsort(colors_lab[:, 0])
sorted_colors_lab = colors_lab[sorted_indices]
sorted_colors_rgb = colors_rgb[sorted_indices]

# Les deux couleurs dominantes de la carapace
dominant_colors_lab_carapace = sorted_colors_lab[:2]
dominant_colors_rgb_carapace = sorted_colors_rgb[:2]

# Affichage des deux couleurs dominantes de la carapace
plt.figure(figsize=(8, 4))

# Diagramme des deux couleurs dominantes de la carapace
plt.subplot(1, 2, 1)
plt.imshow(dominant_colors_rgb_carapace[np.newaxis, :, :].astype(np.uint8))
plt.title("Deux Couleurs Dominantes (Carapace)")
plt.axis("off")

# Image avec le fond supprimé
plt.subplot(1, 2, 2)
plt.imshow(image_no_bg)
plt.title("Image sans Fond")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from skimage import io, color
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create empty lists to store color codes and pixel counts
dominant_colors = []
pixel_counts = []

for k in range(1, 5):
    try:
        # Load the RGB colored image
        image_rgb = io.imread(f'im-{k:03}_im.png')
        image = image_rgb

        # Convert the image to Lab color space
        image_lab = color.rgb2lab(image)

        # Flattening the image for input to the clustering algorithm
        image_flat = image_lab.reshape((-1, 3))

        # Apply K-means clustering to the image
        kmeans = KMeans(n_clusters=5, random_state=42)
        kmeans.fit(image_flat)

        # Get the labels of the clusters and count them
        labels = kmeans.labels_
        counts = np.bincount(labels)

        # Get the colors of the clusters
        cluster_colors = kmeans.cluster_centers_

        # Sort the clusters based on the number of pixels in descending order
        sorted_clusters = np.argsort(counts)[::-1]

        # Iterate over the sorted clusters
        for cluster in sorted_clusters:
            # Get the color code of the dominant color
            dominant_color_code = f"RGB: {cluster_colors[cluster].astype(int)}"

            # Append the dominant color code and pixel count to the lists
            dominant_colors.append(dominant_color_code)
            pixel_counts.append(counts[cluster])

            # Display the dominant color and its color code
            plt.figure(figsize=(8, 4))

            # Dominant color plot
            plt.subplot(1, 2, 1)
            plt.imshow(cluster_colors[cluster][np.newaxis, np.newaxis, :].astype(np.uint8))
            plt.title("Dominant Color")
            plt.axis("off")

            # Color code text
            plt.subplot(1, 2, 2)
            plt.text(0.5, 0.5, dominant_color_code, fontsize=12, ha="center")
            plt.title("Color Code")
            plt.axis("off")

            plt.tight_layout()
            plt.show()

    except FileNotFoundError:
        print(f"Image im-{k:03} not found.")

# Create a DataFrame with the dominant colors and pixel counts
df = pd.DataFrame({"Dominant Color": dominant_colors, "Pixel Count": pixel_counts})
df_sorted = df.sort_values(by="Pixel Count", ascending=False)
print(df_sorted)
```

```python
def plot_color_histogram(image):
    # Convertir l'image en espace de couleurs Lab
    lab_image = color.rgb2lab(image)
    
    # Redimensionner l'image pour accélérer le processus de segmentation
    resized_image = lab_image.reshape(-1, 3)
    
    # Utiliser l'algorithme K-means pour segmenter les couleurs
    kmeans = KMeans(n_clusters=4)
    kmeans.fit(resized_image)
    
    # Récupérer les étiquettes des clusters et les compter
    labels = kmeans.labels_
    counts = np.bincount(labels)
    
    # Trouver les couleurs des clusters
    cluster_colors = [color.lab2rgb([[centroid]])[0][0] for centroid in kmeans.cluster_centers_]
    
    # Trouver les indices des clusters avec les plus grands nombres de pixels (TOP 2)
    top_indices = np.argsort(counts)[-2:]
    
    # Récupérer les couleurs correspondantes aux clusters TOP 2
    top_colors_rgb = [cluster_colors[i] for i in top_indices]
    
    # Convertir les couleurs dominantes de l'espace de couleurs RGB à LAB
    top_colors_lab = [color.rgb2lab([[color_rgb]])[0][0] for color_rgb in top_colors_rgb]
    
    # Créer une figure pour afficher l'histogramme des couleurs
    fig, ax = plt.subplots()
    
    # Afficher l'histogramme des couleurs pour les clusters TOP 2
    ax.bar(top_indices, counts[top_indices], color=top_colors_rgb)
    
    # Définir les étiquettes des axes
    ax.set_xlabel('Cluster')
    ax.set_ylabel('Nombre de pixels')
    
    # Afficher la légende avec les couleurs correspondantes
    legend_labels = ['Couleur dominante 1', 'Couleur dominante 2']
    ax.legend(legend_labels, loc='upper right')
    
    # Afficher la figure
    plt.show()
    
    return top_colors_lab

# Charger l'image de la coccinelle
image = io.imread('im-122_im.png')

# Afficher le TOP 2 des couleurs dominantes en fonction du nombre de pixels et récupérer les couleurs dominantes en LAB
top_colors_lab = plot_color_histogram(image)

# Afficher les valeurs LAB des couleurs dominantes
for i, color_lab in enumerate(top_colors_lab):
    print("Couleur dominante", i+1, "en LAB:", color_lab)
```
